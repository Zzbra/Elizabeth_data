```python
import pandas as pd
from lxml import etree
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

Dans la cellule suivante je cherche une balise xml qui contient la phrase "canine heartworm nematode"

```python
# Load the XML file
with open('aop-wiki-xml-2024-01-01', 'rb') as xml_file:
    tree = etree.parse(xml_file)

# Specify the string you're looking for
search_string = "canine heartworm nematode"

# Use XPath to find elements containing that string in their text
# This XPath expression looks for any element (*) that contains the specified string in its text()
elements_containing_string = tree.xpath(f"//*[contains(text(), '{search_string}')]")

for element in elements_containing_string:
    print(f"Found element: {element.tag}")
    print(f"Text content: {element.text}")
    # You can also navigate to parent/child elements if needed
    # For example: print("Parent element:", element.getparent().tag)
```

Je trouve, la balise s'appelle {http://www.aopkb.org/aop-xml}name

Dans la cellule suivante, j'extrait toutes les informations contenues dans la colonne 'name' de la balise en question

```python
root = tree.getroot()

# Define the namespace map, associating a prefix with the namespace URI
namespaces = {
    'aop': 'http://www.aopkb.org/aop-xml', # Choose the aop namespace
}

# Use the defined prefix in your XPath expression
elements = root.xpath('.//aop:name', namespaces=namespaces)

# Extract data
data_list = []
for element in elements:
    # Extracting text content of each 'name' element found
    data = {'name': element.text}
    data_list.append(data)

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(data_list)

# Display the DataFrame
print(df)
```

Je vérifie ensuite que tout les noms censés êtres présent le sont bien

```python
all_names_set = set(df['name'])
print('human and other cells in culture' in all_names_set)
print('human' in all_names_set)
print('mice' in all_names_set)
print('rat' in all_names_set)
print('canine heartworm nematode' in all_names_set)
print('yeast' in all_names_set)
```
